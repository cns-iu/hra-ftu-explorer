```python
import pandas as pd
import numpy as np
```

```python
kidney = '../ref_data/Anatomogram/ExpDesign-E-CURD-119.tsv'
liver = '../ref_data/Anatomogram/ExpDesign-E-MTAB-10553.tsv'
lung = '../ref_data/Anatomogram/ExpDesign-E-GEOD-130148.tsv'
pancreas = '../ref_data/Anatomogram/ExpDesign-E-MTAB-5061.tsv'
ftu = '../ref_data/FTU Cell Count Table - Cell_Type_Count.csv'
```

```python
kidney_df = pd.read_csv(kidney, sep = '\t')
liver_df = pd.read_csv(liver, sep = '\t')
lung_df = pd.read_csv(lung, sep = '\t')
pancreas_df = pd.read_csv(pancreas, sep = '\t')
ftu_df = pd.read_csv(ftu, sep = '\t')
```

```python
def check_diseased_or_not(df):
    if len(list(df['Sample Characteristic[disease]'].unique())) > 1:
        df = df[df['Sample Characteristic[disease]'] == 'normal']
    return df
        
```

```python
lung_df_modified = check_diseased_or_not(lung_df)
```

```python
# Step 1: Filter the first dataframe based on the "Organ" column
ftu_df['CT ID in CL'] = ftu_df['CT ID in CL'].str.strip()
organ_to_filter = "Lung"  # Replace with the desired organ name
filtered_df1 = ftu_df[ftu_df['Organ'] == organ_to_filter]

lung_df_modified['Transformed Ontology Term'] = lung_df_modified['Factor Value Ontology Term[inferred cell type - ontology labels]'].str.extract(r'CL_(\d+)')
lung_df_modified['Transformed Ontology Term'] = 'CL:' + lung_df_modified['Transformed Ontology Term'].str.strip()

lung_df_modified.head()
```

```python
# Merge data to add 'FTU Label in Uberon' based on the match
lung_df_modified['FTU'] = lung_df_modified.merge(
    filtered_df1[['CT ID in CL', 'FTU Label in Uberon']],
    how='left',
    left_on='Transformed Ontology Term',
    right_on='CT ID in CL'
)['FTU Label in Uberon']

print(lung_df_modified.head(2))
```
