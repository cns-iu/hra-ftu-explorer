    3. Any dataset that is been shared among two FTUs

```python
# Load the JSON files
version = 'v0.11.1'
base_url = 'https://raw.githubusercontent.com/x-atlas-consortia/hra-pop/main/output-data/'
atlas_enriched_dataset_graph_url = f'{base_url}{version}/atlas-enriched-dataset-graph.jsonld'
atlas_as_cell_summaries_url = f'{base_url}{version}/atlas-as-cell-summaries.jsonld'
crosswalks_url = 'https://raw.githubusercontent.com/hubmapconsortium/hra-workflows-runner/main/crosswalking-tables/crosswalks.jsonld'
ftu_cell_count_path = '../ref_data/FTU Cell Count Table - Cell_Type_Count.csv'
sankey_url = f'{base_url}{version}/reports/universe-ad-hoc/sankey.csv'


ftu_cell_count = pd.read_csv(ftu_cell_count_path, sep='\t')

# Define fetch_csv function
def fetch_csv(url):
    response = requests.get(url, verify=certifi.where())
    if response.status_code == 200:
        from io import StringIO
        return pd.read_csv(StringIO(response.text))
    else:
        print(f"Error fetching {url}: Status code {response.status_code}")
        return None
    
def fetch_json(url):
    response = requests.get(url, verify=certifi.where())
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error fetching {url}: Status code {response.status_code}")
        return None

# Load the Sankey data
sankey = fetch_csv(sankey_url)

# Load the JSON data
crosswalks_url_graph = fetch_json(crosswalks_url)
atlas_enriched_dataset_graph = fetch_json(atlas_enriched_dataset_graph_url)
atlas_as_cell_summaries = fetch_json(atlas_as_cell_summaries_url)
```

```python
# Extract the UBERON IDs from the organ_to_uberon dictionary
uberon_ids = set(pd.Series(ftu_cell_count['Organ_ID'].unique()).str.split('/').str[-1].str.replace('_', ':'))

# Extract the unique Organ_IDs from popv, celltypist, and azimuth
popv_ids = set(crosswalk_df[crosswalk_df['tool'] == 'popv']['organ_id'].unique())

celltypist_ids = set(crosswalk_df[crosswalk_df['tool'] == 'celltypist']['organ_id'].unique())

azimuth_ids = set(crosswalk_df[crosswalk_df['tool'] == 'azimuth']['organ_id'].unique())

# Find common organ IDs between organ_to_uberon and the different datasets
common_popv = uberon_ids.intersection(popv_ids)
common_celltypist = uberon_ids.intersection(celltypist_ids)
common_azimuth = uberon_ids.intersection(azimuth_ids)

print('Organs in popv:', list(common_popv))
print('\n')
print('Organs in celltypist:', list(common_celltypist))
print('\n')
print('Organs in azimuth:', list(common_azimuth))
```

```python
# Rename 'CT ID in CL' to 'CL_ID' for consistency
ftu_cell_count = ftu_cell_count.rename(columns={'CT ID in CL': 'CL_ID'})
ftu_cell_count = ftu_cell_count.rename(columns={'CT Label in CL': 'CL_Label_FTU'})

# Filter FTU cell count to include only rows where `CT ID in CL` contains `CL`
ftu_cell_count_filtered = ftu_cell_count[ftu_cell_count['CL_ID'].str.contains('CL', na=False)]

# Select necessary columns from ftu_cell_count
ftu_cell_count_filtered = ftu_cell_count_filtered[['Organ', 'Organ_ID', 'FTU Label in Uberon', 'FTU ID in Uberon', 'CL_ID', 'CL_Label_FTU']]

# Merge the dataframes based on 'CL_ID' (from ftu_cell_count_filtered) and 'cell_id' (from crosswalk_df)
merged_df = ftu_cell_count_filtered.merge(
    crosswalk_df.loc[crosswalk_df['tool'] == 'azimuth', ['organ_id', 'organ_level', 'cell_id']],
    left_on='CL_ID',  # Column in ftu_cell_count_filtered
    right_on='cell_id',  # Column in crosswalk_df
    how='left',
    suffixes=('', '_azimuth')
)

# Rename 'cell_label' to 'CL_Label_azimuth' if it exists
if 'cell_label' in merged_df.columns:
    merged_df = merged_df.rename(columns={'cell_label': 'CL_Label_azimuth'})

# Merge with celltypist data
merged_df = merged_df.merge(
    crosswalk_df.loc[crosswalk_df['tool'] == 'celltypist', ['organ_id', 'organ_level', 'cell_id']],
    left_on='CL_ID',
    right_on='cell_id',
    how='left',
    suffixes=('', '_celltypist')
)

if 'cell_label' in merged_df.columns:
    merged_df = merged_df.rename(columns={'cell_label': 'CL_Label_celltypist'})

# Merge with popv data
merged_df = merged_df.merge(
    crosswalk_df.loc[crosswalk_df['tool'] == 'popv', ['organ_id', 'organ_level', 'cell_id']],
    left_on='CL_ID',
    right_on='cell_id',
    how='left',
    suffixes=('', '_popv')
)

if 'cell_label' in merged_df.columns:
    merged_df = merged_df.rename(columns={'cell_label': 'CL_Label_popv'})

# Populating the columns azimuth, celltypist, popv with 1 if there is a match, otherwise 0
merged_df['azimuth'] = merged_df.get('CL_Label_azimuth', pd.Series(dtype='int')).notna().astype(int)
merged_df['celltypist'] = merged_df.get('CL_Label_celltypist', pd.Series(dtype='int')).notna().astype(int)
merged_df['popv'] = merged_df.get('CL_Label_popv', pd.Series(dtype='int')).notna().astype(int)

# Renaming 'CL_Label_FTU' to 'CT Label'
merged_df = merged_df.rename(columns={'CL_Label_FTU': 'CT Label'})

# Selecting columns for final output
final_columns = ['Organ', 'Organ_ID', 'FTU Label in Uberon', 'FTU ID in Uberon', 'CL_ID', 'CT Label', 'azimuth', 'celltypist', 'popv']
ftu_ct_ann_info_df = merged_df[final_columns]

# Removing any duplicates
ftu_ct_ann_info_az_df = ftu_ct_ann_info_df.drop_duplicates( 
    subset=['CL_ID', 'azimuth'], keep='last').reset_index(drop=True)

ftu_ct_ann_info_ct_df = ftu_ct_ann_info_df.drop_duplicates( 
    subset=['CL_ID', 'celltypist'], keep='last').reset_index(drop=True)

ftu_ct_ann_info_pv_df = ftu_ct_ann_info_df.drop_duplicates( 
    subset=['CL_ID', 'popv'], keep='last').reset_index(drop=True)

# Final creation of dataframe
ftu_ct_ann_info_df = ftu_ct_ann_info_pv_df

# Update azimuth, celltypist, and popv flags
ftu_ct_ann_info_df['azimuth'] = ftu_ct_ann_info_df['CL_ID'].isin(azimuth_ids).astype(int)
ftu_ct_ann_info_df['celltypist'] = ftu_ct_ann_info_df['CL_ID'].isin(celltypist_ids).astype(int)
ftu_ct_ann_info_df['popv'] = ftu_ct_ann_info_df['CL_ID'].isin(popv_ids).astype(int)


# Save the final dataframe to a CSV file
output_path = '../output/FTU-CT-AnnTool-info.csv'
ftu_ct_ann_info_pv_df.to_csv(output_path, index=False)

# Display the first few rows of the final dataframe
print(ftu_ct_ann_info_pv_df.head())
print(ftu_ct_ann_info_pv_df.shape)
```

```python
def swarmplot_cell_types_per_tool_level(df, output_path='swarmplot_cell_types_per_tool_level.png'):
    # Prepare the data
    annotation_columns = ['popv', 'celltypist_L1', 'celltypist_pkl', 'azimuth_L1', 'azimuth_L2', 'azimuth_L3', 'azimuth_L4', 'azimuth_L5']
    # Create a new column combining organ and cell type information
    df['Organ_Cell_Type'] = df['Organ'] + ' - ' + df['CL_Label_FTU']
    
    # Reshape the data for plotting
    data = df.melt(id_vars=['Organ_Cell_Type'], value_vars=annotation_columns, 
                   var_name='Annotation Tool Level', value_name='Match')
    
    # Filter matched rows
    data = data[data['Match'] == 1]

    # Plot using Seaborn's catplot
    g = sns.catplot(
        data=data,
        x='Annotation Tool Level',
        y='Organ_Cell_Type',
        hue='Annotation Tool Level',
        kind='swarm',
        height=10,
        aspect=2,
        palette='Set2'
    )

    g.set_titles('Cell Types Per Annotation Tool Level (Including Organ Information)')
    g.set_axis_labels('Annotation Tool Level', 'Organ - Cell Types')
    plt.xticks(rotation=45)
    plt.tight_layout()

    # Save the plot as a high-quality file
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    plt.show()

# Call the updated function and save the plot
swarmplot_cell_types_per_tool_level(final_df, output_path='../output/figures/high_quality_swarmplot.png')
```

```python
def heatmap_cell_types_per_tool_level(df, output_path='heatmap_cell_types_per_tool_level.png'):
    # Prepare the data
    annotation_columns = ['popv', 'celltypist_L1', 'celltypist_pkl', 'azimuth_L1', 'azimuth_L2', 'azimuth_L3', 'azimuth_L4', 'azimuth_L5']
    # Add organ information to cell types
    df['Organ_Cell_Type'] = df['Organ'] + ' - ' + df['CL_Label_FTU']
    
    # Reshape the data for the heatmap
    data = df.melt(id_vars=['Organ_Cell_Type'], value_vars=annotation_columns, 
                   var_name='Annotation Tool Level', value_name='Match')
    
    # Filter matched rows
    data = data[data['Match'] == 1]

    # Pivot table for heatmap
    heatmap_data = data.pivot_table(index='Organ_Cell_Type', columns='Annotation Tool Level', aggfunc='size', fill_value=0)

    # Plot heatmap
    plt.figure(figsize=(14, 10))
    sns.heatmap(heatmap_data, cmap='YlGnBu', linewidths=0.5, annot=True, fmt='d')
    plt.title('Cell Types Per Annotation Tool Level (Including Organ Information)', fontsize=14)
    plt.xlabel('Annotation Tool Level', fontsize=12)
    plt.ylabel('Organ - Cell Types', fontsize=12)
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.tight_layout()

    # Save the heatmap as a high-quality image
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    plt.show()

# Call the updated function and save the heatmap
heatmap_cell_types_per_tool_level(final_df, output_path='../output/figures/high_quality_heatmap.png')
```

#### Understanding the structure, content, and formatting of the JSON files namely, atlas-as-cell-summaries.jsonld and atlas-enriched-dataset-graph.jsonld.

```python
# Function to extract structure and keys of the JSON
def extract_structure(data, level=0):
    structure = {}
    if isinstance(data, dict):
        for key, value in data.items():
            structure[key] = extract_structure(value, level + 1)
    elif isinstance(data, list) and len(data) > 0:
        structure = [extract_structure(data[0], level + 1)]
    else:
        structure = None
    return structure 
```

#### Extract gene information for idenfied CT - dataset combination from cell-summary.

```python
# Define the function to create the DataFrame
def create_dataframe_with_demographics(json_data):
    data = []

    # Iterate through the entries in the JSON graph
    for entry in json_data["@graph"]:
        sex = entry.get('sex', 'n/a')
        race = entry.get('race', 'n/a')
        bmi = entry.get('bmi', 'n/a')
        
        for sample in entry.get('samples', []):
            rui_location = sample.get('rui_location', {})
            all_collisions = rui_location.get('all_collisions', [])
            sample_datasets = sample.get('datasets', [])
            
            x_value = rui_location.get('x_dimension', 'n/a')
            y_value = rui_location.get('y_dimension', 'n/a')
            z_value = rui_location.get('z_dimension', 'n/a')
            
            for collision in all_collisions:
                collisions = collision.get('collisions', [])
                for col in collisions:
                    reference_organ = col.get('reference_organ', 'n/a')
                    # Simplify "reference_organ" value
                    if reference_organ.startswith("http://purl.org/ccf/latest/ccf.owl#"):
                        reference_organ = reference_organ.split("#")[-1].replace("VHF", "").replace("VHM", "")

                    as_label = col.get('as_label', 'n/a')

                    for dataset in sample_datasets:
                        summaries = dataset.get('summaries', [])
                        dataset_id = dataset.get('@id', 'n/a')
                        link = dataset.get('link', 'n/a')

                        for summary in summaries:
                            annotation_method = summary.get('annotation_method', 'n/a')
                            for summ in summary.get('summary', []):
                                if isinstance(summ, dict):
                                    cell_id = summ.get('cell_id', 'n/a')
                                    cell_label = summ.get('cell_label', 'n/a')

                                    for gene_expr in summ.get('gene_expr', []):
                                        if isinstance(gene_expr, dict):
                                            gene_id = gene_expr.get('gene_id', 'n/a')
                                            gene_label = gene_expr.get('gene_label', 'n/a')
                                            ensembl_id = gene_expr.get('ensembl_id', 'n/a')
                                            mean_gene_expr_value = gene_expr.get('mean_gene_expr_value', 'n/a')

                                            row = {
                                                'reference_organ': reference_organ,
                                                'organ_id': dataset.get('organ_id', 'n/a'),
                                                'cell_id': cell_id,
                                                'cell_label': cell_label,
                                                'annotation_method': annotation_method,
                                                'as_label': as_label,
                                                'sex': sex,
                                                'race': race,
                                                'bmi': bmi,
                                                'x_value': x_value,
                                                'y_value': y_value,
                                                'z_value': z_value,
                                                'link': link,
                                                'dataset': dataset_id,
                                                'gene_id': gene_id,
                                                'gene_label': gene_label,
                                                'ensembl_id': ensembl_id,
                                                'mean_gene_expr_value': mean_gene_expr_value
                                            }
                                            data.append(row)

    # Convert the collected data into a DataFrame
    df = pd.DataFrame(data)
    return df

# Create the DataFrame with modified "reference_organ" logic
df_atlas_enriched = create_dataframe_with_demographics(atlas_enriched_dataset_graph)
df_atlas_enriched.head()
```

### Understand the JSON files

```python
# Function to extract as_label, link, and cell_id from the JSON data
def extract_all_entries(json_data):
    result = []
    for entry in json_data['@graph']:
        for sample in entry.get('samples', []):
            rui_location = sample.get('rui_location', {})
            all_collisions = rui_location.get('all_collisions', [])
            for collision in all_collisions:
                collisions = collision.get('collisions', [])
                for col in collisions:
                    as_label = col.get('as_label', 'unknown')
                    for section in sample.get('sections', []):
                        for dataset in section.get('datasets', []):
                            link = dataset.get('link', 'unknown')
                            for summary in dataset.get('summaries', []):
                                for summary_item in summary.get('summary', []):
                                    cell_id = summary_item.get('cell_id', 'unknown')
                                    cell_label = summary_item.get('cell_label', 'unknown')
                                    new_row = {
                                        'as_label': as_label,
                                        'link': link,
                                        'cell_label': cell_label,
                                        'cell_id': cell_id
                                    }
                                    result.append(new_row)
    return result

# Extract all entries
extracted_data_enriched = extract_all_entries(json_data_enriched)

# Convert extracted data to a DataFrame
extracted_df_enriched = pd.DataFrame(extracted_data_enriched)

# Display the DataFrame's head
print(extracted_df_enriched.head())
```

```python
# Extract cell IDs from the summary section
def extract_cell_ids(data):
    cell_ids = []
    for item in data['@graph']:
        samples = item.get("samples", [])
        for sample in samples:
            sections = sample.get("sections", [])
            for section in sections:
                datasets = section.get("datasets", [])
                for dataset in datasets:
                    summaries = dataset.get("summaries", [])               
                    for summary in summaries:
                        sum_details = summary.get("summary", [])
                        for sum_detail in sum_details:
                            cell_id = sum_detail.get("cell_id")
                            cell_ids.append(cell_id)
    return cell_ids

cell_ids = extract_cell_ids(json_data_enriched)
print(f"Cell IDs from the file: {len(cell_ids)}")
```
